```python
column=df.columns.tolist()
```

```python
column=[c for c in column if c not in ['Class']]
target='Class'
```

```python
import numpy as np
```

```python
randoms=np.random.RandomState(50)
```

```python
x=df[column]
y=df[target]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score
```

```python
isolation=IsolationForest( n_estimators=100,
    max_samples=len(x),
    contamination=outlier_fraction,
    random_state=randoms,
    verbose=0)
```

```python
isolation.fit(x)
scores_prediction = isolation.decision_function(x)
y_pred = isolation.predict(x)
```

```python
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1
n_errors = (y_pred != y).sum()
```

```python
score=accuracy_score(y,y_pred)
score
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof=LocalOutlierFactor(n_neighbors=20,
    algorithm='auto',
    leaf_size=30,
    metric='minkowski',
    p=2,
    metric_params=None,
    contamination=outlier_fraction)
```

```python
y_pred=lof.fit_predict(x)
scores_prediction = lof.negative_outlier_factor_
```

```python
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1
n_errors = (y_pred != y).sum()
score=accuracy_score(y,y_pred)
score
```
